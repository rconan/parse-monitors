```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri
plt.style.use('dark_background')
```

```python
cases = np.load("gir_fxy_max.pkl",allow_pickle=True)
cfd = [x[0] for x in cases]
fx = [x[1] for x in cases]
fy = [x[2] for x in cases]
```

```python
fig,ax = plt.subplots(figsize=(14,6))
w = [2,7,12,17]
for l in range(3):
    a0 = l*20
    for k in range(4):
        if l>0:
            ax.fill_betweenx([0,300],a0+k*5,a0+(k+1)*5,color=f"C{k+2}",alpha=0.5*(l+1)/3)
        else:
            ax.fill_betweenx([0,300],a0+k*5,a0+(k+1)*5,color=f"C{k+2}",alpha=0.5,label=f"{w[k]}m/s")
ax.stem(range(60),fx,label="Fx")
ax.stem(range(60),fy,linefmt='C1',markerfmt='oC1',label="Fy")
ax.grid()
ax.set_ylabel("GIR max. force [N]")
ax.set_xlabel("CFD cases #")
ax.set_xticks(range(0,61,5));
ax.legend()
ax.set_title("Elevations: 90 -> 60 -> 30 deg.")
```
